### Data Visualization With Seaborn

Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and infromative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
```

```python
## Basic plotting with seaborn

tips = sns.load_dataset('tips')
tips
```

```python
### Create a scatter plot

import matplotlib.pyplot as plt

plt.scatter(x='total_bill',y='tip',data=tips)

plt.title('Scatter plot of Total bill vs tip')

plt.show()
```

```python
## lINE PLOT

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Line Plot of total bill by size')
plt.show()
```

```python
### Categorical variables Plots
## Bar Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of total bill by day')
plt.show()
```

```python
## Box plot

sns.boxplot(x='day',y='total_bill',data=tips)

plt.show()
```

```python
### Violin plot

sns.violinplot(x='day',y='total_bill',data=tips)

```

```python
## Histograms

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
### KDE plot

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
## Pair Plot

sns.pairplot(tips)
```

```python
### Heatmap

corr = tips[['total_bill','tip','size']].corr()

corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
